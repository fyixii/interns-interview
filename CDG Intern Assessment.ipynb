# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing data file
df = pd.read_excel('10 Years Financial Data.xlsx',header=1)
```

```python
df.head()
```

```python
#Reassigning column names
df.columns=["Profit & Loss (SGD '000)","Graph","Dec-03","Dec-04","Dec-05","Dec-06","Dec-07","Dec-08","Dec-09","Dec-10","Dec-11","Dec-12","Dec-13","Dec-14","Dec-15","Dec-16","Dec-17","Dec-18","Dec-19"]
```

```python
df.head()
```

```python
#Dropping the first row as its a duplicate
df.drop([0],inplace=True)
```

```python
#Dropping the graph column as it doesnt show in jupyter notebook
df2 = df.drop(columns=['Graph'])
```

```python
df2.head()
```

```python
#Transposed data frame to reflect P&L attribute as columns and years as rows
df_transposed = df2.transpose()
```

```python
df_transposed.head()
```

```python
#Reassigning of headers
header = df_transposed.iloc[0]
df_transposed.columns = header
```

```python
df_transposed.head()
```

```python
df_transposed.drop(index="Profit & Loss (SGD '000)",inplace=True)
```

```python
df_transposed.head()
```

```python
#Replacing the "n.a" & "-" with the standard numpy "NaN" to indicate null values
df_transposed.replace(['n.a.','-'],np.nan,inplace=True)
```

```python
#This is the clean dataset that we can work with
df_transposed
```

```python
df_transposed.shape
```

```python
df_transposed.describe()
```

```python
df_transposed.info()
```

```python
#Number of empty columns (P&L attributes) by rows (year)
df_transposed.isnull().sum(axis=1)
```

```python
#Number of empty rows (year) by columns (P&L attributes)
df_transposed.isnull().sum(axis = 0)
```
